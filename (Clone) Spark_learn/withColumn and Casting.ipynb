```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv")\
    .option("header",True)\
    .option("inferSchema",True)\
    .load("/Volumes/sparkcatalog/raw/source/raw_orders/rawData.csv")
```

```python
df.show()
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import expr, coalesce, lit

df_f = df.withColumn(
    "amount",
    coalesce(expr("try_cast(amount as int)"), lit(0))
)


df_f.show()
```

```python
df_n = df_f.sort(["id","age"],ascending=[0,1])
df_n.show()
```

```python
df_n.printSchema()
```

```python
df_n = df_n.limit(1)
df_n.show()
```

```python
df_d = df_n.drop("amount")
df_d.show()
```
