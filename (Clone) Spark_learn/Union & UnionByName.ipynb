```python
data1 = [(1,"Jone",30),(2,"Mary",20),(3,"Bob",25),(4,"Mike",35),(5,"Tom",40)]
df1 = spark.createDataFrame(data1,['id','name','age'])

data2 = [(6,"Jone",30),(7,"Mary",20),(8,"Bob",25),(9,"Mike",35),(10,"Tom",40)]
df2 = spark.createDataFrame(data2,['id','name','age'])
```

```python
union = df1.union(df2).dropDuplicates(["name"])
union.show()
```

```python
data3 = [(30,"mick",11),(40,"tom",12),(50,"jack",13),(60,"jane",14),(70,"mary",15)]
df3 = spark.createDataFrame(data3,['age','name','id'])
unionByName = df1.union(df2).unionByName(df3).drop("age")
unionByName.show()
```
