#Partitions

```python
data = [
    (1, "Laptop", "India", 50000),
    (2, "Mobile", "India", 20000),
    (3, "TV", "USA", 40000),
    (4, "Watch", "USA", 10000),
    (5, "Shoes", "UK", 5000),
    (6, "Bag", "UK", 3000)
]

columns = ["order_id", "product", "country", "price"]

df = spark.createDataFrame(data, columns)

display(df)
```

```python
df.write.format("delta") \
    .mode("overwrite") \
    .save("/mnt/demo/orders")
```

#orders/
   part-0000...
   part-0001...

###write with partitions


```python
df.write.format("delta") \
    .partitionBy("country") \
    .mode("overwrite") \
    .save("/mnt/demo/orders_partition")
```
