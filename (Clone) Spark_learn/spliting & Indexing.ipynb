```python
from pyspark.sql.functions import * 
from pyspark.sql.types import *
```

```python
data = [("388/3 , pattatharasiAmman kovil street , Aathupplayam , Tiruppur , Tamilnadu")]

df = spark.createDataFrame(data,["Adress"])

df.show()
```

```python
df_split = df.withColumn("Door Num",split("Adress",",")[0])\
    .withColumn("Street",split("Adress",",")[1])\
    .withColumn("Village",split("Adress",",")[2])\
    .withColumn("City",split("Adress",",")[3])\
    .withColumn("Distrc",split("Adress",",")[4])
```

```python
df_split = df_split.select(["Door Num","Street","Village","City","Distrc"])
```

```python
df_split.show()
```

#EXPLODE

```python
data = [("388/3 , pattatharasiAmman kovil street , Aathupplayam , Tiruppur , Tamilnadu")]

df_new = spark.createDataFrame(data,["Adress"])
```

```python
df_new.show()
```

```python
df_new = df_new.withColumn("Adress_explod",split("Adress",","))
df_new = df_new.withColumn("explod",explode("Adress_explod"))
display(df_new)
```

# Array_contains


```python
df_new_df = df_new.withColumn("city1",array_contains("Adress_explod","Tiruppur"))
display(df_new_df)
```
