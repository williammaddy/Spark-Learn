#Handling Nulls

```python
from pyspark.sql.functions import initcap
from pyspark.sql.types import *
```

```python
data = [(1,'jhon',None),(2,'Ben10','ab'),(3,'ChottaBheem','Dolakpore'),(4,'spiderMan','newYork'),(None,None,None),(2,'Ben10','ab'),(5,'shinChan',None),(6,None,None),(7,None,'Peak')]

df = spark.createDataFrame(data,['id','name','address'])
df = df.withColumn("Name",initcap('name'))
df.show()
```

```python
df = df.dropDuplicates()
df.show()
```

```python
df.show()
```

## DROP ALL NULL

```python
df_all = df.dropna("all")
df_all.show()
```

```python
df_any = df.dropna("any")
df_any.show()
```

```python
df_subset = df.dropna(subset=["Name","address"])
df_subset.show()
```

```python
df_fillna = df.fillna({'id':'0','Name':'No Name Found','address':'Adress cant found'})
df_fillna.show()
```
