```python
from pyspark.sql.functions import floor, rand, concat_ws, col
```

```python
sales_data = [
    (101, "Laptop", 50000),
    (101, "Mouse", 800),
    (101, "Keyboard", 1500),
    (101, "Monitor", 12000),
    (101, "Headphone", 2500),
    (101, "Tablet", 18000),
    (101, "Printer", 9000),
    (101, "SSD", 4500),
    (101, "RAM", 3200),
    (101, "Camera", 25000),
    (101, "Phone", 35000),
    (101, "Speaker", 2000),
    (102, "Laptop", 48000),
    (103, "Keyboard", 1800),
    (104, "Mouse", 900),
    (105, "Printer", 8500)
]

sales = spark.createDataFrame(
    sales_data,
    ["customer_id", "product", "amount"]
)

sales.show()
```

```python
salted_df = sales.withColumn("salt",floor(rand()*4))
```

```python
from pyspark.sql.functions import sum

partial_result = salted_df.groupBy(
    "customer_id",
    "salt"
).agg(
    sum("amount").alias("partial_total")
)

partial_result.show()
```

```python
partial_result = partial_result.groupBy(
    "customer_id",
).agg(
    sum("partial_total").alias("total")
)
```
