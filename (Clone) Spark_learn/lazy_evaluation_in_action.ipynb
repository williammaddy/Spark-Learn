**This is a lazy Evaluation**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
Data = [
    (1,"manikandan","mani@gmail.com"),
    (2,"geetha","geetha@gmail.com"),
    (3,"karthika","karthika@gmail.com"),
    (4,"lakshmanan","laksh@gmail.com")
]

column = ["id","name","email"]

df = spark.createDataFrame(Data,column)

df= df.select("id","email")
df=df.filter(col("id")>1)
```
