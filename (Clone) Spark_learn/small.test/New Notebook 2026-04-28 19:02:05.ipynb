#collect__list

```python
from pyspark.sql.functions import * 
from pyspark.sql.types import *
```

```python
data = [
    ("A", 10),
    ("A", 20),
    ("A", 10),
    ("B", 30),
    ("B", 40)
]

df = spark.createDataFrame(data, ["category", "value"])
df.show()
```

```python
df_csv = df.groupBy("category").agg(collect_list("value").alias("values"))
df_csv.show()
```

#PIVOT

```python
data = [
    ("A", "Jan", 100),
    ("A", "Feb", 200),
    ("B", "Jan", 300),
    ("B", "Feb", 400)
]

df = spark.createDataFrame(data, ["name", "month", "sales"])
df.show()
```

```python
df = df.groupBy("name").pivot("month").agg(sum("sales"))
```

```python
display(df)
```

#WHEN AND OTHERWISE

```python
df_if = df.withColumn("result", when(col("Jan")==100,"Mid").otherwise("High"))
df_if.show()
```
