```python
from pyspark.sql.functions import when , col , expr
from pyspark.sql.types import * 
```

```python
df = spark.read.format("csv")\
    .option("header",True)\
    .option("inferSchema",True)\
    .load("/Volumes/sparkcatalog/raw/source/raw_orders/rawData.csv")
```

```python
df.show()
df.printSchema()
```

```python
df = df.fillna({"age":"0"})
df.show()
```

```python
df_cast = df.withColumn("age",expr("try_cast(age as int)"))\
    .withColumn("amount",expr("try_cast(amount as int)"))
```
