```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv")\
    .option("header",True)\
    .option("inferSchema",True)\
    .load("/Volumes/sparkcatalog/raw/source/raw_orders/rawData.csv")
```

```python
display(df)
```

```python
df_s = df.select("id","name","age","amount")
df_f = df_s.filter(col("name") == "Mani")
```

```python
df_r = df_f.withColumnRenamed("name","username")
```

```python
display(df_r)
```

```python
df_n = df_r.withColumn("with_tax",df_r.amount * 1.05)
display(df_n)
```
