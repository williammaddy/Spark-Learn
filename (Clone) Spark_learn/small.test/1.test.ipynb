```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv")\
    .option("header",True)\
    .option("mod","PERMISSIVE")\
    .load("/Volumes/sparkcatalog/raw/source/raw_orders/rawData.csv")
    
df.show()
```

```python
df = df.select("id","name","age")
df.show()
```

```python
df = df.withColumn("age",col("age").cast("int"))
df_f = df.filter(col("age")<30)
display(df_f)
```
