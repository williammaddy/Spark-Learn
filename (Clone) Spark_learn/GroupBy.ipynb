```python
from pyspark.sql.functions import *
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

data = [
    (1, "Mani", "IT", ["Python", "SQL"]),
    (2, "Kumar", "HR", ["Excel", "Communication"]),
    (3, "Ravi", "IT", ["Python", "Spark"]),
    (4, "Anu", "Finance", ["Excel", "Accounting"]),
    (5, "Divya", "IT", ["Spark", "Scala"]),
    (6, "Arun", "HR", ["Communication", "Excel"]),
    (7, "Priya", "Finance", ["Accounting", "SQL"])
]

columns = ["id", "name", "department", "skills"]

df_f = spark.createDataFrame(data, columns)
df_f.show(truncate=False)
```

# Array_contains

```python
df = df_f.filter(array_contains(df_f.skills,"Python"))

df.show()
```

# GroupBy


```python
df_g = df_f.groupBy("department").agg(count("*").alias("total"))
df_g.show()
```

# approx_count_distinct

```python
df_a = df_f.select(approx_count_distinct("name").alias("Approx_count")).show()
```

#Collect_list

```python
df_c = df_f.groupBy("department")\
    .agg(collect_list("skills").alias("employee"))
df_c.select("employee").show()
```

```python
data = spark.read.format("csv")\
    .option("header",True)\
    .option("inferSchema",True)\
    .load("/Volumes/sparkcatalog/raw/source/raw_orders/orders.csv")
display(data)
```

#PIVOTING

```python
data_pi = data.groupBy("customer_id").pivot("order_status")\
    .agg(count("order_id"))
data_pi = data_pi.fillna(0)
display(data_pi)
```

```python

df_if = data_pi.withColumn("flag",when(col("Returned")==1,"low")\
    .when(col("Returned")<3,"Mid").when(col("Returned")>3,"Hight"))
```

```python
df_if = df_if.filter(col("flag")=="Hight")
df_if.show()
```
