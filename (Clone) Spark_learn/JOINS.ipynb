```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

df1 = spark.createDataFrame([
    (1, "Alice"),
    (2, "Bob"),
    (3, "Cathy"),
    (4, "Dan")
], ["id", "name"])

df2 = spark.createDataFrame([
    (1, "HR"),
    (2, "IT"),
    (4, "Finance")
], ["id", "dept"])

df1.show()
df2.show()

# Inner Join
df_joined = df1.join(df2, on="id", how="inner")
df_joined.show()
df_joined.explain()
```

```python
departments = spark.createDataFrame([
    (1, "HR"),
    (2, "IT"),
    (5, "Finance")
], ["emp_id", "dept"])
```

```python
employees = spark.createDataFrame([
    (1, "Alice"),
    (2, "Bob"),
    (3, "Cathy"),
    (4, "David")
], ["emp_id", "name"])
```

```python
df_j = employees.join(departments, on="emp_id", how="left")

df_j.show()
```

#INNER JOIN

```python
customers_data = [
(1,"Arun","Chennai"),
(2,"Bala","Coimbatore"),
(3,"Charan","Madurai"),
(4,"Divya","Chennai"),
(5,"Eswar","Trichy"),
(6,"Fathima","Salem"),
(7,"Gokul","Chennai"),
(8,"Hari","Madurai")
]

customers_df = spark.createDataFrame(customers_data,["customer_id","name","city"])
```

```python
orders_data = [
(101,1,"2024-01-01",500),
(102,2,"2024-01-02",700),
(103,3,"2024-01-03",300),
(104,1,"2024-01-04",900),
(105,4,"2024-01-05",400),
(106,7,"2024-01-06",1000),
(107,10,"2024-01-07",600)  # invalid customer_id (for filtering)
]

orders_df = spark.createDataFrame(orders_data,["order_id","customer_id","order_date","amount"])
```

```python
df_f = customers_df.join(orders_df, on="customer_id",how="inner")
```

```python
df_f.show()
```
